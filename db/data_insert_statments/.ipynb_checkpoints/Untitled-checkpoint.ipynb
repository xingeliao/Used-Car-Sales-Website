```python
import pandas as pd
```

```python
df = pd.read_csv('parts.tsv', sep='\t')
```

```python
df
```

```python
trimmed_df = df[['VIN', 'order_num', 'vendor_name', 'price', 'qty']].copy()
trimmed_df.head()
```

```python
trimmed_df.loc[:,'subtotal'] = trimmed_df['price'] * trimmed_df['qty']
trimmed_df.head()
```

```python
grouped = trimmed_df.groupby(['VIN', 'order_num', 'vendor_name']).agg({'subtotal': 'sum'}).reset_index()
```

```python
grouped['order_num'] = grouped['order_num'].astype('str')
grouped.dtypes
```

```python
grouped.loc[:, 'purchase_order_number'] = grouped['VIN'] + '-' + grouped['order_num'].str.pad(width=3, fillchar='0')
```

```python
grouped.head()
```

```python
grouped.to_csv('parts_with_calculated_cost.csv', index=False)
```
